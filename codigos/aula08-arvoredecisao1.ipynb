```python
X,y = iris.data, iris.target
```

```python
classificador = tree.DecisionTreeClassifier() # instancio o modelo da rvore de deciso
classificador.fit(X,y)#treino do modelo
fig = plt.figure(figsize=(30,25)) # defino o tamanho da figura
tree.plot_tree(classificador,feature_names = iris.feature_names,class_names=iris.target_names.tolist(),
               filled= True)
plt.savefig('arvore1.jpg')
```
